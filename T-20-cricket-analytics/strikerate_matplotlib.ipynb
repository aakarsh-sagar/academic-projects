Batting plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_bat = pd.read_excel('Batting.xlsx')
df_bat
```

```python
# filter out only the batters
# batters_only = df_bat.loc[df_bat['Specialization'].isin(['Batter', 'Batsman']), 'SR']
bat_career = df_bat.loc[(df_bat['Specialization'].isin(['Batter', 'Batsman'])) &(df_bat['YEAR']== 'Career')]
batters_only = df_bat.loc[(df_bat['Specialization'].isin(['Batter', 'Batsman'])) & (df_bat['YEAR']!= 'Career')]
# batters_only = df_bat.loc[df_bat['Specialization'].isin(['Batter', 'Batsman'])]
batters_only.SR.replace('-',float(0)).sort_values(ascending=False)

```

```python
batters_only

# batters_only.to_excel('batonly.xlsx', index=0)
```

```python

bat_career[['SR', 'AVG']] = bat_career[['SR', 'AVG']].replace('-', 0.0).astype(float)

bat_career.to_excel('batting_career.xlsx', index=0)
bat_career

```

```python
# plot
# career SR
career_SR_top = bat_career.nlargest(10, 'SR')
career_SR_top
plt.bar(career_SR_top['Name'], career_SR_top['SR'])
plt.xlabel('Player Name')
plt.ylabel('Strike Rate')
plt.title('Top 10 Strike Rates by Career')
plt.xticks(rotation=45)
plt.show()
```

```python
# Career Average for batters



career_avg_top = bat_career.nlargest(10, 'AVG')
career_avg_top
plt.bar(career_avg_top['Name'], career_avg_top['AVG'])
plt.xlabel('Player Name')
plt.ylabel('Batting Average')
plt.title('Top 10 Batting Averages by Career')
plt.xticks(rotation=45)
plt.show()



```

```python
# Career sixes top 10

career_six_top = bat_career.nlargest(10, '6S')
career_six_top
plt.bar(career_six_top['Name'], career_six_top['6S'])
plt.xlabel('Player Name')
plt.ylabel('Sixes scored')
plt.title('Top 10 Players who scored most sixes by Career')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Runs scored

career_runs_top = bat_career.nlargest(10, 'RUNS')
career_runs_top
plt.bar(career_runs_top['Name'], career_runs_top['RUNS'])
plt.xlabel('Player Name')
plt.ylabel('Runs scored')
plt.title('Top 10 Players who scored most Runs by IPL Career')
plt.xticks(rotation=45)
plt.show()
```

```python
# Wicketkeeper figures by Career
wk_career = df_bat.loc[(df_bat['Specialization'].isin(['Wicketkeeper Batter', 'WK-Batsman','Wicket-keeper','Wicketkeeper'])) &(df_bat['YEAR']== 'Career')]

career_ct_top = wk_career.nlargest(10, 'CT')

plt.bar(career_ct_top['Name'], career_ct_top['CT'])
plt.xlabel('Player Name')
plt.ylabel('Catches taken')
plt.title('Top 10 Wicketkeepers who secured most catches by IPL Career')
plt.xticks(rotation=45)
plt.show()


```

```python
career_st_top = wk_career.nlargest(10, 'ST')

plt.bar(career_st_top['Name'], career_st_top['ST'])
plt.xlabel('Player Name')
plt.ylabel('Stumpings')
plt.title('Top 10 Wicketkeepers who secured most stumpings by IPL Career')
plt.xticks(rotation=45)
plt.show()
```

```python
# Dismissals by the wicketkeepers
wk_career.loc[: , 'Dismissals'] = wk_career['ST'] + wk_career['CT']
wk_career.loc[: , 'avg_dismissals'] = wk_career['Dismissals'] / wk_career['MAT']

wk_dism_top = wk_career.nlargest(10, 'Dismissals')
plt.bar(wk_dism_top['Name'], wk_dism_top['Dismissals'])
plt.xlabel('Player Name')
plt.ylabel('Dismissals')
plt.title('Top 10 Wicketkeepers who secured most Dismissals by IPL Career')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average dismissals
wk_dism_avg = wk_career.nlargest(10, 'avg_dismissals')
plt.bar(wk_dism_avg['Name'], wk_dism_avg['avg_dismissals'])
plt.xlabel('Player Name')
plt.ylabel('Average Dismissals')
plt.title('Top 10 Wicketkeepers with highest average dismissals by IPL Career')
plt.xticks(rotation=45)
plt.show()
```

Bowler stuff

```python
df_bowl = pd.read_excel('Bowling_Update.xlsx')
bowl_career = df_bowl.loc[(df_bowl['Specialization'].isin(['Bowler'])) &(df_bowl['YEAR.1']== 'Career')]
bowl_career.loc[:,['AVE', 'SR.1']] = bowl_career[['AVE', 'SR.1']].replace('-', 0.0).astype(float)
bowl_career
```

```python
#plotting the lowest economy
bowl_career['ECON'] = bowl_career['ECON'].astype(float)
bowl_econ_top =  bowl_career.nsmallest(10, 'ECON')

plt.bar(bowl_econ_top['Name'], bowl_econ_top['ECON'])
plt.xlabel('Player Name')
plt.ylabel('Economy')
plt.title('Top 10 Bowlers who secured the lowest economies IPL Career')
plt.xticks(rotation=45)
plt.show()
```

```python
# strike rate
bowl_career['SR.1'] = bowl_career['SR.1'].astype(float)

bowl_sr_top =  bowl_career.nsmallest(11, 'SR.1')

plt.bar(bowl_sr_top['Name'][1:11], bowl_sr_top['SR.1'][1:11])
plt.xlabel('Player Name')
plt.ylabel('Strike Rate')
plt.title('Top 10 Bowlers who secured the lowest strike rates in their IPL Career')
plt.xticks(rotation=45)
plt.show()
```

```python
# total number of wickets
bowl_wkts_top =  bowl_career.nlargest(10, 'WKTS')

plt.bar(bowl_wkts_top['Name'], bowl_wkts_top['WKTS'],color = 'purple')
plt.xlabel('Player Name')
plt.ylabel('Wickets')
plt.title('Top 10 Bowlers who secured the most wickets in their IPL Career')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average wickets
bowl_career.loc[:,'AVG_WKTS'] = bowl_career['WKTS']/bowl_career['MAT.1']
bowl_avgwkts_top =  bowl_career.nlargest(10, 'AVG_WKTS')
plt.bar(bowl_avgwkts_top['Name'], bowl_avgwkts_top['AVG_WKTS'],color = 'darkgreen')
plt.xlabel('Player Name')
plt.ylabel('Average Wickets per game')
plt.title('Top 10 Bowlers with highest average in wickets in their IPL Career')
plt.xticks(rotation=45)
plt.show()
```
