```python
import numpy as np
import pandas as pd
```

```python
path = "PlayerData.xlsx"
table = pd.read_excel(path)
df = table.copy()
```

## Deleting duplicate header column

```python
df.drop(0, inplace = True)
```

## Create a copy of df to performs checks and verify the edits before doing them in the original
## Check the nulls

```python
df_copy = df.copy()
df_copy.isnull().sum()
```

## Performing backward fill on names first, as there are a few orphan stats, this seems to belong to players above the nulls

```python
df_copy = df.copy()
df_copy.Name.fillna(method = "ffill", inplace = True)
li = ['Country','IPL Debut','Specialization','DOB']
df_copy.loc[:,li] = df_copy.loc[:, li].fillna(method="ffill")
df_copy.isnull().sum()
df_copy[(df_copy['Name'] == "Ishant Sharma")] # Example check
```

### Results come right, however at this point we notice that the Batting and bowling figures are not aligned by the year in the file. Hence, we decide to split the files based on spcializations. So that the analysis i sdone right.

```python
df.isnull().sum()
```

```python
df.Name.fillna(method = "ffill", inplace = True)
li = ['Country','IPL Debut','Specialization','DOB']
df.loc[:,li] = df.loc[:, li].fillna(method="ffill")
df.isnull().sum()
df[(df['Name'] == "Ishant Sharma")]
```

## Using the DOB to calculate age of players that maybe used for further analysis

```python
def age(x):
    return (pd.to_datetime('today').normalize() - x)
```

```python
df_bat['Age'] = df_bat['DOB'].apply(age).astype('<m8[Y]')
```

```python
df_bat
```

## Saving Progress

```python
df_bat.to_excel('Batting.xlsx',index=False)
```

## Testing a function that I'll apply

```python
def highscore(a):
    container = ""
    if isinstance(a, (int, float)):
        return int(a)
    else:
        for i in a:
            if i != "*":
                container += i
        return int(container)
```

```python
df_bat['HighScore'] = df_bat['HS'].apply(highscore)
df_bat['out-notout'] = df_bat['HS'].apply(lambda x: 'out' if isinstance(x,(int,float)) else 'notout')
df_bat
```

## Saving Progress

```python
df_bat.to_excel('Batting.xlsx',index = False)
```

```python
def clean(x):
    if len(x) == 2:
        return '-'
    elif len(x) == 1:
        return (x[0])
    elif len(x) == 3 and int(x[0]) < 2000:
        return str(x[1] + "/" + x[0][2:])
    else:
        return (x[1] + "/" + x[2])
```

```python
df_bowl['BBM Figures'] = df_bowl['New Col'].apply(clean)
```
